```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df
```

```python
def paid_more(row):
    fare = row["Fare"]
    median = df["Fare"].median()

    if fare > median:
        return 1
    else:
        return 0


df["paid_more_than_median"] = df.apply(paid_more, axis=1)
```

```python
df
```

```python
count_more = df["paid_more_than_median"].sum()
total = len(df)

print("Passengers who paid more than median:", count_more)
print("Total passengers:", total)
```

```python
young = df[df["Age"] < 30]["Survived"].value_counts()
old   = df[df["Age"] >= 30]["Survived"].value_counts()
```
